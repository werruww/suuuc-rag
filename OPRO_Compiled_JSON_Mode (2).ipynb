````python
import dspy
import requests

class OllamaPredictor(dspy.Predict):
    def __init__(self, model="llama3"):
        super().__init__()
        self.model = model

    def forward(self, prompt):
        response = requests.post(
            "http://localhost:11434/api/generate",
            json={"model": self.model, "prompt": prompt, "stream": False}
        ).json()
        return [response["response"]]

ollama_lm = OllamaPredictor("llama3")
dspy.settings.configure(lm=ollama_lm)
print(ollama_lm("Explain vector databases."))
````

````python
class GenerateResponse(dspy.Signature):
    """Follow the task_instructions (Input Field) and generate the response (Output Field) according to the output format given by response_format (Input Field). You will be given references from (Task-Specific Input Field).
VERY IMPORTANT!! ONLY OUTPUT JSON AND NOTHING ELSE!! THIS IS EXTEMELY IMPORTANT AND YOUR OUTPUT WILL BE USELESS OTHERWISE! DO NOT START YOUR OUTPUT WITH ```json OR END IT WITH ```! ONLY OUTPUT THE KEYS GIVEN BY THE response_format (Input Field)!!
    """

    task_instructions = dspy.InputField(desc="(Input Field)")
    response_format = dspy.InputField(desc="(Input Field)")
    references = dspy.InputField(desc="Task-Specific Input Field")
    response = dspy.OutputField(desc="(Output Field)")
````

# Load JSON Metric

````python
import json

def is_valid_json_output(output):
    try:
        parsed = json.loads(output)
        return isinstance(parsed.get("answer"), str) and isinstance(parsed.get("confidence"), int) and 0 <= parsed["confidence"] <= 5
    except json.JSONDecodeError:
        return False

# Could put types here -- would love to test one prompt for all types eventually
def JSON_Metric(gold, pred, trace=None):
    return is_valid_json_output(pred.response)
````

````python
import json

# ✅ 1. تحميل بيانات JSON من الملف
filename = "/content/structured-rag/data/wiki-answerable-questions.json"  # غيّر الاسم إن كان مختلفًا
with open(filename, "r", encoding="utf-8") as f:
    data = json.load(f)

# ✅ 2. استخراج العناوين أو الأسئلة كقائمة
titles = [item["answerable_question"] for item in data]
contexts = [item["abstract"] for item in data]

print(f"✅ تم تحميل {len(titles)} عنصر.\n")

# 🔁 3. وظيفة البحث التفاعلية
def interactive_search():
    while True:
        query = input("\n🔍 اكتب جزءاً من السؤال أو العنوان (أو 'خروج' للخروج): ").strip().lower()
        if query in ["خروج", "exit", "quit"]:
            print("👋 وداعاً!")
            break

        # العثور على العناصر المتطابقة جزئياً
        matches = [(idx, t) for idx, t in enumerate(titles) if query in t.lower()]
        if not matches:
            print("❌ لا توجد نتائج للبحث.")
            continue

        # عرض النتائج المحتملة
        print("\n📝 نتائج البحث:")
        for idx, title in matches:
            print(f"{idx}: {title}")

        # طلب اختيار العنصر
        sel = input("\n🔢 اختر رقم العنصر لعرض النص (أو 'رجوع'): ").strip().lower()
        if sel in ["رجوع", "back"]:
            continue

        try:
            idx = int(sel)
            if idx < 0 or idx >= len(contexts):
                print("❌ رقم غير صالح.")
                continue
        except ValueError:
            print("❌ إدخال غير صالح.")
            continue

        # عرض المحتوى بالكامل
        print("\n📌 النص المقابل:")
        print(contexts[idx])
        print("\n" + "-"*50)

# ✅ 4. تشغيل الوظيفة
interactive_search()
````

### شغال يعرض الاسءلة او العناوين ف الداتا لتختار منها

````python
import json
import os

# ✨ قم بتعديل اسم الملف هنا إذا أردت استخدام ملف مختلف
FILENAME = "/content/structured-rag/data/wiki-answerable-questions.json"

if not os.path.exists(FILENAME):
    raise FileNotFoundError(f"❌ الملف '{FILENAME}' غير موجود في المسار الحالي.")

# 1. تحميل بيانات JSON
with open(FILENAME, "r", encoding="utf-8") as f:
    data = json.load(f)

# 2. استخراج العناوين (أسئلة) والنصوص (contexts)
titles = []
contexts = []
for item in data:
    # تعديل المفاتيح حسب هيكل ملفك
    title = item.get("answerable_question") or item.get("question") or "<بدون عنوان>"
    context = item.get("abstract") or item.get("context") or ""
    titles.append(title)
    contexts.append(context)

print(f"✅ تم تحميل {len(titles)} عنوانًا من الملف.\n")

# 3. عرض القائمة والتفاعل
def interactive_display():
    while True:
        print("\n📚 هذه بعض العناوين (الأول 20):")
        for i, t in enumerate(titles[:20]):
            print(f"{i}: {t}")
        print("...")

        sel = input("\n🔢 اختر رقم العنوان (أو 'استعراض' للبحث بالكلمة، أو 'خروج' للخروج): ").strip().lower()
        if sel in ["خروج", "exit", "quit"]:
            print("👋 انتهى. إلى اللقاء.")
            break
        elif sel == "استعراض":
            query = input("🔍 اكتب كلمة للبحث في العناوين: ").strip().lower()
            matches = [(i, t) for i, t in enumerate(titles) if query in t.lower()]
            if not matches:
                print("❌ لا توجد نتائج.")
                continue
            print("\n📝 نتائج البحث:")
            for i, t in matches[:20]:
                print(f"{i}: {t}")
            continue
        else:
            try:
                idx = int(sel)
                if idx < 0 or idx >= len(contexts):
                    print("❌ الرقم خارج النطاق.")
                    continue
            except:
                print("❌ إدخال غير صحيح.")
                continue

            print("\n📌 النص المرتبط:")
            print(contexts[idx])
            print("\n" + "-"*60)

# 4. تشغيل الوظيفة
interactive_display()
````
